```python
import numpy as np
import mglearn
import matplotlib.pyplot as plt 
from ML_tool import file_tool
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
```

```python
data_file = file_tool.input_file('./forge_dataset.txt')
data_file.read_from_file()
raw_x, raw_y = data_file.sort_forge_dataset()

raw_x = np.array(raw_x)
raw_y = np.array(raw_y)

print(raw_x.shape)
print(raw_y.shape)
```

```python
clf1 = LogisticRegression().fit(raw_x, raw_y)
clf2 = LinearSVC(max_iter=100000).fit(raw_x, raw_y)

fig, axes = plt.subplots(1, 3, figsize=(10, 3))
for cv, ax in zip([0.01, 1, 100], axes):
    clf = LogisticRegression(C=cv).fit(raw_x, raw_y)
    mglearn.plots.plot_2d_separator(clf, raw_x, fill=False, eps=0.5, ax=ax, alpha=0.7)
    mglearn.discrete_scatter(raw_x[:, 0], raw_x[:, 1], raw_y, ax=ax)
    ax.set_title(f"C={cv:.2f}")
    ax.set_xlabel("Feature 0")
    ax.set_ylabel('Feature 1')

plt.show()
```
